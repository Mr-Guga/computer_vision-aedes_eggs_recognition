```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

plt.rcParams['figure.dpi'] = 200
```

```python
img = cv.imread('../assets/images/aedes_template.jpg', 0)
img_blur = cv.blur(img, (7, 7))

canny_edge = cv.Canny(img, 200, 300)

plt.subplot(2, 2, 1), plt.imshow(img,cmap = 'gray')
plt.title('original'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2), plt.imshow(img_blur, cmap = 'gray')
plt.title('blur'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3), plt.imshow(canny_edge,cmap = 'gray')
plt.title('edge'), plt.xticks([]), plt.yticks([])

plt.show()
```
